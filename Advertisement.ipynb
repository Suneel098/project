## Description

The scatter plots show the **relationship between different advertising channels and Sales**. Each graph represents one advertising medium (TV, Newspaper, Radio, Social Media, and Promotion Video) plotted against Sales.

The **colored points** represent the actual data values, showing how much sales occurred for different advertisement spending levels. The **black line represents the regression line**, which indicates the predicted linear relationship between advertisement spending and sales.

By analyzing these graphs, we can observe whether increasing the spending on a particular advertising channel leads to an **increase or decrease in sales**. If the regression line slopes upward, it indicates a **positive relationship**, meaning sales increase with higher advertising spending. If the line slopes downward, it indicates a **negative relationship**, meaning the advertisement channel may not effectively increase sales.

Overall, these visualizations help identify **which advertising channels contribute more effectively to sales performance** and support better decision-making in marketing strategies.
