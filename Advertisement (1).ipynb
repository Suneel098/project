Insights from Regression Formulas
The calculated regression formulas reveal the linear relationship between each advertising channel's spending and 'Sales'.

Tv: Sales = 0.01 * Tv + 34.47
NewsPaper: Sales = -0.07 *
NewsPaper + 35.78
Radio: Sales = 0.02 * Radio + 34.50
SocialMedia: Sales = 0.01 * SocialMedia + 34.54
Prmotionvideo: Sales = -0.04 * Prmotionvideo + 35.46

```python
advertising_channels = ['Tv', 'NewsPaper', 'Radio', 'SocialMedia', 'Prmotionvideo']

regression_results = {
    'Tv': {'slope': 0.01, 'intercept': 34.47},
    'NewsPaper': {'slope': -0.07, 'intercept': 35.78},
    'Radio': {'slope': 0.02, 'intercept': 34.50},
    'SocialMedia': {'slope': 0.01, 'intercept': 34.54},
    'Prmotionvideo': {'slope': -0.04, 'intercept': 35.46}
}

colors = ['red','green','blue','purple','orange']

for i, channel in enumerate(advertising_channels):

    plt.figure(figsize=(6,4))

    sns.scatterplot(x=advertisement_df[channel],
                    y=advertisement_df['Sales'],
                    color=colors[i])

    # Get regression parameters before using them in the f-string
    slope = regression_results[channel]['slope']
    intercept = regression_results[channel]['intercept']

    # Formula text - now correctly terminated and uses defined variables
    formula_text = f'Sales = {slope:.2f} * {channel} + {intercept:.2f}'

    plt.plot(advertisement_df[channel],
             slope*advertisement_df[channel] + intercept,
             color='black')

    plt.title(f"{channel} vs Sales")
    plt.xlabel(channel)
    plt.ylabel("Sales")

    plt.show()
```

Linear Regression Description

The above code uses Linear Regression visualization to analyze the relationship between advertising spending and Sales. For each advertising channel (TV, Newspaper, Radio, Social Media, and Promotion Video), a scatter plot is created to display the distribution of data points.

In the graph, the colored dots represent the actual observations, showing how sales vary with different levels of advertisement spending. The black line represents the regression line, which is the best-fit line that summarizes the linear relationship between the advertising variable and sales.

Linear regression helps determine how strongly an advertising channel influences sales. If the regression line slopes upward, it indicates a positive relationship, meaning that higher advertisement spending may lead to increased sales. If the line slopes downward, it indicates a negative relationship, suggesting that increasing spending on that channel may not improve sales.

By examining these graphs for each advertising channel, we can understand which marketing platforms have a stronger impact on sales performance and use this information to make better marketing and budgeting decisions.
